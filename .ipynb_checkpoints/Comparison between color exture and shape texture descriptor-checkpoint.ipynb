## 1. Texture Feature

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
# from google.colab.patches import cv2_imshow
from skimage import data
from skimage.color import rgb2gray
```

```python
# import the necessary packages For Local Binary Pattern
from skimage import feature
from skimage import io
import numpy as np
```

#### - Read Images

```python

img1 = cv2.imread('tringle.png')
img2 = cv2.imread('square.png')
img3 = cv2.imread('circle.png')


plt.imshow(img1)
plt.axis('off')  # Hide axis values
plt.show()
plt.imshow(img2)
plt.axis('off')  # Hide axis values
plt.show()
plt.imshow(img3)
plt.axis('off')  # Hide axis values
plt.show()

# Wait for a key press and then close the window
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

#### - Convert Images

```python

gray1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img3,cv2.COLOR_BGR2GRAY)
print('Gray Scale Image')
plt.imshow(gray1)
plt.axis('off')  # Hide axis values
plt.show()
plt.imshow(gray2)
plt.axis('off')  # Hide axis values
plt.show()
plt.imshow(gray3)
plt.axis('off')  # Hide axis values
plt.show()
```

```python
# Define radius, epsilon and number of points
numPoints = 16
radius = 2
eps = 1e-7

lbp1 = feature.local_binary_pattern(gray1, numPoints, radius, method="uniform")
lbp2 = feature.local_binary_pattern(gray2, numPoints, radius, method="uniform")
lbp3 = feature.local_binary_pattern(gray3, numPoints, radius, method="uniform")


(hist1, h1) = np.histogram(lbp1.ravel(),bins=np.arange(0, numPoints + 3),range=(0, numPoints + 2))
#normalisasi vektor histogram
hist1 = hist1.astype("float")
hist1 /= (hist1.sum() + eps)
plt.bar(h1[:-1], hist1, width = 0.5, color='#0504aa',alpha=0.7)
plt.title('LBP Histogram for Triangle Feature',fontsize=15)
```

```python
(hist2, h2) = np.histogram(lbp2.ravel(),bins=np.arange(0, numPoints + 3),range=(0, numPoints + 2))
# normalisasi vektor histogram
hist2 = hist2.astype("float")
hist2 /= (hist2.sum() + eps)
plt.bar(h2[:-1], hist2, width = 0.5, color='#0504aa',alpha=0.7)
plt.title('LBP Histogram for Circle Feature',fontsize=15)
```

```python

(hist3, h3) = np.histogram(lbp3.ravel(),bins=np.arange(0, numPoints + 3),range=(0, numPoints + 2))
# normalisasi vektor histogram
hist3 = hist3.astype("float")
hist3 /= (hist3.sum() + eps)
plt.bar(h3[:-1], hist3, width = 0.5, color='#0504aa',alpha=0.7)
plt.title('LBP Histogram for Square Feature',fontsize=15)
```

```python
print(' Feature Extraction for Triangle  ==> ',hist1)

print(' Feature Extraction for Circle ==> ',hist2)

print(' Feature Extraction for Square ==> ',hist3)
```

## Perbandingan Vector Ciri LBP Ketiga Gambar(Segitiga,Lingkaran dan Persegi)

```python
from scipy.spatial import distance
#  Perbandingan Ketiga Vektor Ciri Menggunakan minkowski distance

# Jarak antara Vektor Segitiga dan Lingkaran
jarakSeg_Lin = distance.minkowski(hist1,hist2)
print('Jarak antara Vektor Segitiga dan Lingkaran adalah ', jarakSeg_Lin)

# Jarak antara Vektor Segitiga dan Persegi
jarakSeg_Per = distance.minkowski(hist1,hist3)
print('Jarak antara Vektor Segitiga dan Persegi adalah ', jarakSeg_Per)

# Jarak antara Vektor Lingkaran dan Persegi
jarakLin_Per = distance.minkowski(hist2,hist3)
print('Jarak antara Vektor  Lingkaran dan Persegi adalah ', jarakLin_Per)
```

#  2. Color Feature

```python

# MendapatKan Ciri Vektor Mean dan Standar Deviasi Untuk Warna blue, green, dan  red pada gambar segitiga, Lingkaran dan Persegi

(meanSeg, stdSeg) = cv2.meanStdDev(img1)
(meanLin, stdLin) = cv2.meanStdDev(img2)
(meanPer, stdPer) = cv2.meanStdDev(img3)

```

```python
# Concatenate Statistic Color Feature Mean and Standart Deviation

stat_segitiga = np.concatenate([meanSeg, stdSeg]).flatten()
stat_lingkaran = np.concatenate([meanLin, stdLin]).flatten()
stat_persegi = np.concatenate([meanPer, stdPer]).flatten()

print('Vektor Statistik Segitiga : ',stat_segitiga)
print('Vektor Statistik Lingkaran : ',stat_lingkaran)
print('Vektor Statistik Persegi : ',stat_persegi)
```

## Perbandingan Vector Ciri Warna Ketiga Gambar(Segitiga,Lingkaran dan Persegi)

```python
from scipy.spatial import distance
#  Perbandingan Ketiga Vektor Ciri Menggunakan minkowski distance

# Jarak antara Vektor Segitiga dan Lingkaran
jarakSeg_Lin = distance.minkowski(stat_segitiga,stat_lingkaran)
print('Jarak antara Vektor Segitiga dan Lingkaran adalah ', jarakSeg_Lin)

# Jarak antara Vektor Segitiga dan Persegi
jarakSeg_Per = distance.minkowski(stat_segitiga,stat_persegi)
print('Jarak antara Vektor Segitiga dan Persegi adalah ', jarakSeg_Per)

# Jarak antara Vektor Lingkaran dan Persegi
jarakLin_Per = distance.minkowski(stat_lingkaran,stat_persegi)
print('Jarak antara Vektor  Lingkaran dan Persegi adalah ', jarakLin_Per)
```

# 3. Shape Feature

```python

## Canny Edge Detection unutk segitga, lingkaran dan persegi

edgeSegitiga = cv2.Canny(gray1,100,200)
edgeLingkaran = cv2.Canny(gray2,100,200)
edgePersegi = cv2.Canny(gray3, 100,200)

plt.subplots(1),plt.imshow(edgeSegitiga,cmap = 'gray')
plt.title('Edge Image Segitiga'), plt.xticks([]), plt.yticks([])

plt.subplots(1),plt.imshow(edgeLingkaran,cmap = 'gray')
plt.title('Edge Image Lingkaran'), plt.xticks([]), plt.yticks([])

plt.subplots(1),plt.imshow(edgePersegi,cmap = 'gray')
plt.title('Edge Image Persegi'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
## Proses mendapatkan contour pada setiap gambar(segitiga, lingkaran dan persegi)

ret1,thresh1 = cv2.threshold(edgeSegitiga,127,255,0)
contours1,hierarchy1 = cv2.findContours(thresh1, 1, 2)

ret2,thresh2 = cv2.threshold(edgeLingkaran,127,255,0)
contours2,hierarchy2 = cv2.findContours(thresh2, 1, 2)

ret3,thresh3 = cv2.threshold(edgePersegi,127,255,0)
contours3,hierarchy3 = cv2.findContours(thresh3, 1, 2)

# Centroid  untuk Segitiga, Lingkaran dan Persegi
cnt1 = contours1[0]
cnt2 = contours2[0]
cnt3 = contours3[0]
```

### Mendapatkan Fitur Shape menggunakan Contour Approximation Vector

```python
epsilon1 = 0.1*cv2.arcLength(cnt1,True)
approx1 = cv2.approxPolyDP(cnt1,epsilon1,True)

epsilon2 = 0.1*cv2.arcLength(cnt2,True)
approx2 = cv2.approxPolyDP(cnt2,epsilon2,True)

epsilon3 = 0.1*cv2.arcLength(cnt3,True)
approx3 = cv2.approxPolyDP(cnt3,epsilon3,True)
```

## Perbandingan Vector Ciri Shape Ketiga Gambar(Segitiga,Lingkaran dan Persegi)

```python
# Perbandingan Ketiga Vektor Ciri Menggunakan Euclidean Distance

from scipy.spatial import distance

# Ubah Matriks manjadi Vector dengan satu dimensi
feat_Seg =  approx1.ravel()
feat_Ling =  approx2.ravel()
feat_Per =  approx3.ravel()

# Ambil maximum Lenght
ml = max(len(feat_Seg),len(feat_Ling),len(feat_Per))
# Tambahkan nol jika panjang vector kurang dari maximum length
feat_Seg1 = np.concatenate([feat_Seg , np.zeros(ml-len(feat_Seg))])
feat_Ling1 = np.concatenate([feat_Ling , np.zeros(ml-len(feat_Ling))])
feat_Per1 =  np.concatenate([feat_Per, np.zeros(ml-len(feat_Per))])

# Jarak antara Vektor Segitiga dan Lingkaran
jarakSeg_Lin = distance.euclidean(feat_Seg1 ,feat_Ling1)
print('Jarak antara Vektor Segitiga dan Lingkaran adalah ', jarakSeg_Lin)

# Jarak antara Vektor Segitiga dan Persegi
jarakSeg_Per = distance.euclidean(feat_Seg1,feat_Per1)
print('Jarak antara Vektor Segitiga dan Persegi adalah ', jarakSeg_Per)

# Jarak antara Vektor Lingkaran dan Persegi
jarakLin_Per = distance.euclidean(feat_Ling1,feat_Per1)
print('Jarak antara Vektor  Lingkaran dan Persegi adalah ', jarakLin_Per)
```
